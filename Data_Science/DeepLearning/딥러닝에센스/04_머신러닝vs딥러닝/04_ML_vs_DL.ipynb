```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold

import os
import random
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
np.random.seed(2024)
random.seed(2024)
os.environ['PYHONHASHSEED'] = 'O'
```

```python
data = pd.read_csv('/content/drive/MyDrive/인공지능/Part8. 딥러닝 에센스/data/Imbalance_data/creditcard.csv')
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
print('Class 0:', data['Class'].value_counts()[0]) # No frauds
print('Class 1:', data['Class'].value_counts()[1]) # frauds
```

```python
data['Class'].value_counts()[0] / (data['Class'].value_counts()[0] + data['Class'].value_counts()[1]) * 100
```

```python
from sklearn.preprocessing import RobustScaler
rob_scaler = RobustScaler()

for col in ['Amount', 'Time']:
  data[col] = rob_scaler.fit_transform(data[col].values.reshape(-1, 1))
```

```python
data.columns
```

```python
# Undersample
data1 = data.sample(frac=1)
fraud_data = data1.loc[data1['Class']==1]
no_fraud_data = data1.loc[data1['Class']==0][:492]
```

```python
len(fraud_data), len(no_fraud_data)
```

```python
balanced_data = pd.concat([fraud_data, no_fraud_data])
balanced_data = balanced_data.sample(frac=1)
```

```python
balanced_data.head()
```

```python
x = balanced_data.drop('Class', axis=1)
y = balanced_data['Class']
```

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=2024)
```

```python
train_x = train_x.values
train_y = train_y.values

test_x = test_x.values
test_y = test_y.values
```

```python
lr = LogisticRegression()
lr.fit(train_x, train_y)
```

```python
from sklearn.model_selection import cross_val_score

cross_val_score(lr, train_x, train_y, cv=5)
```

```python
y_pred = lr.predict(test_x)
from sklearn.metrics import accuracy_score
accuracy_score(test_y, y_pred)
```

```python
# SMOTE
X = data.drop('Class', axis = 1)
y = data['Class']
```

```python
X.shape
```

```python
kfold = StratifiedKFold(n_splits=5)
X_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2024)

X_train = X_train.values
y_train = y_train.values

x_test = x_test.values
y_test = y_test.values
```

```python
print(X_train.shape, x_test.shape)
```

```python
from imblearn.over_sampling import SMOTE
lr = LogisticRegression()
for train, test in kfold.split(X_train, y_train):
  smote = SMOTE(random_state=2024)
  X_train_smote, y_train_smote = smote.fit_resample(X_train[train], y_train[train])

  lr.fit(X_train_smote, y_train_smote)
```

```python
y_pred = lr.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

# 3.CIFAR10 이미지 분류

## 1) ML

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets
```

```python
(x_train, y_train), (x_test, y_test) = datasets.cifar10.load_data()
```

```python
x_train = x_train.astype('float32')
x_test = x_test.astype('float32')
```

```python
x_train = x_train/255.
x_test = x_test/255.
```

```python
flat_x_train = x_train.reshape(x_train.shape[0], -1)
flat_x_test = x_test.reshape(x_test.shape[0], -1)

print(flat_x_train.shape, flat_x_test.shape)
```

```python
y_train = y_train.reshape(y_train.shape[0])
y_test = y_test.reshape(y_test.shape[0])
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
# KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(flat_x_train, y_train)
```

```python
y_pred_knn = knn.predict(flat_x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred_knn)
```

```python
rf = RandomForestClassifier(n_estimators=100, criterion='entropy', max_depth=70)
rf.fit(flat_x_train, y_train)
```

```python
y_pred_rf = rf.predict(flat_x_test)
accuracy_score(y_test, y_pred_rf)
```

```python
dt = DecisionTreeClassifier(criterion='entropy', max_depth=100)
dt.fit(flat_x_train, y_train)
```

```python
y_pred_dt = dt.predict(flat_x_test)
accuracy_score(y_test, y_pred_dt)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=500)
pca.fit(flat_x_train)
```

```python
x_train_pca = pca.transform(flat_x_train)
x_test_pca = pca.transform(flat_x_test)
print(x_train.shape, x_test_pca.shape)
```

```python
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train_pca, y_train)

y_pred_knn = knn.predict(x_test_pca)
accuracy_score(y_test, y_pred_knn)
```

```python
rf = RandomForestClassifier(n_estimators=100, criterion='entropy', max_depth=70)
rf.fit(x_train_pca, y_train)

y_pred_rf = rf.predict(x_test_pca)
accuracy_score
```

```python
dt = DecisionTreeClassifier(criterion='entropy', max_depth=100)
dt.fit(x_train_pca, y_train)

y_pred_dt = dt.predict(x_test_pca)
accuracy_score(y_test, y_pred_dt)
```

## 2) DL

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import datasets, Model, layers, optimizers
```

```python
(images_train, labels_train), (images_test, labels_test) = datasets.cifar10.load_data()
```

```python
print(images_train.shape, labels_train.shape)
print(images_test.shape, labels_test.shape)
```

```python
images_train = images_train.astype('float32') / 255.
images_test = images_test.astype('float32') / 255.
```

```python
input_shape = (32, 32, 3)
num_classes = 10
```

```python
model = keras.models.Sequential([
    layers.Conv2D(32, (3, 3), padding='same', activation='relu',input_shape=input_shape),
    layers.MaxPool2D(strides=(2,2)),
    layers.BatchNormalization(),

    layers.Conv2D(64, (3,3), padding='same', activation='relu'),
    layers.MaxPool2D(strides=(2,2)),
    layers.BatchNormalization(),

    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.BatchNormalization(),
    layers.Dense(num_classes, activation='softmax')
])
```

```python
model.compile(optimizer=optimizers.Adam(learning_rate=0.001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model_train = model.fit(images_train, labels_train, validation_split=0.2, batch_size=128, epochs=3, shuffle=True, verbose=2)
```

```python
test_lost, test_acc = model.evaluate(images_test, labels_test, verbose=2)
print('test accuracy:', test_acc)
```
