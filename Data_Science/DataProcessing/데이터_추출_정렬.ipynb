# 조건에 맞는 데이터 추출

```python
# a가 3 이상인 데이터를 출력
df[df['a'] >= 3]
```

```python
# a가 3이상인 데이터 중 a, c열만 출력
df[df['a']>=3][['a','c']]
```

```python
# a가 3이상이고, b가 16미만인 데이터 출력
df[(df['a']>=3) & (df['b']<16)] 
```

```python
# a가 3이하 이거나 7이상인 데이터 출력
df[(df['a']<=3) | (df['b']>=7)]
```

```python
# a가 3이상이고, b가 16미만이거나 c가 30인 데이터를 출력하시오.
df[(df['a'] >=3) & ((df['b'] < 16) | (df['c'] == 30))]
```

# 정렬

```python
import pandas as pd

df = pd.DataFrame({'a' : [2, 3, 2, 7, 4], 'b' : [2, 1, 3, 5, 3], 'c' : [1, 1, 2, 3, 5]})

# 내리차순 # 결과를 저장하고 싶으면 inplace = True
df.sort_index(ascending = False, inplace = True)
df
```

```python
# 인덱스 초기화 reset_index
df = df.reset_index() # drop = True -> index : None
df
```

```python
# 내리차순 # 결과를 저장하고 싶으면 inplace = True
df.sort_index(ascending = True, inplace = True)
df
```

```python
# a열 기준으로 오름차순정렬
df.sort_values(by=['a'], inplace = True)
df
```

```python
# a, b열 기준으로 오름차순으로 정렬
df.sort_values(by=['a','b'])
df
```

```python
# a열 기준으로 오름차순 정렬  한 이후, b열 기준으로 내림차순 정렬
df.sort_values(by = ['a','b'], ascending = [True, False], inplace = True)
df
```

# 그룹화

```python
import pandas as pd

df1 = pd.DataFrame({'ID' : [1, 2, 3, 4, 5], '가입일' : ['2021-01-02', '2021-01-04', '2021-01-10', '2021-02-10', '2021-02-24'], '성별' : ['F', 'M', 'F', 'M', 'M']})
df2 = pd.DataFrame({'구매순서' : [1, 2, 3, 4, 5], 'ID' : [1, 1, 2, 4, 1], '구매월' : [1, 1, 2, 2, 3], '금액' : [1000, 1500, 2000, 3000, 4000]})

# df1은 회원 정보를 저장한 데이터 프레임이며, df2는 각 회원의 구매 내역을 저장한 데이터 프레임이다. 
# 각 회원의 정보와 구매 내역을 취합하여 하나의 데이터 프레임으로 만드시오.

pd.merge(df1, df2, how ='left', on='ID')
```

```python
# df1 은 회원의 정보를 저장한 데이터 프레임이며, df2는 각회원의 구매 내역을 저장한 데이터 프레임이다.
# 각 회원의 누적 금액을 회원 Id별로 구하시오.

s2 = df2.groupby(by = ['ID'])['금액'].sum()
pd.merge(df1, s2, how = 'left', on = 'ID')
```

```python
# 각 회원의 월별 누적 금액을 회원 ID별로 구하시오.
s3 = df2.groupby(by = ['ID', '구매월'])['금액'].sum().reset_index()
pd.merge(df1, s3, how = 'left', on = 'ID')
```

```python
df2
```

```python
# 각 회원의 최대 사용 금액/최소 사용금액과 최저 수수료의 값을 구하시오.
s4 = df2.groupby(by = ['ID']).agg({'금액' : [max, min]})
s4.reset_index()
s4.columns
```

# 피벗테이블

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'가입월' : [1, 1, 1, 2, 2, 3], '탈퇴월' : [1, 2, 3, 2, 3, 3], '탈퇴회원수' : [101, 52, 30, 120, 60, 130]})
pivot = pd.pivot_table(df, values = '탈퇴회원수', index = ['가입월'], columns = ['탈퇴월'],fill_value = 0)
```

```python
pivot
```

```python
# 과일 매장의 판매내역 data set
import random
a = []
b = []
for i in range(100):
	a.append(random.randint(1,3))
	b.append(random.randint(1,3))

df = pd.DataFrame({'품목' : a, '크기' : b})

df['금액'] = df['품목'] * df['크기'] * 500
df['수수료'] = df['금액'] * 0.1
df.head()
```

```python
fruit_name = {1 : '토마토', 2 : '바나나', 3 : '사과'}
fruit_size = {1 : '소', 2 : '중', 3 : '대'}

df['품목'] = df['품목'].map(fruit_name)
df['크기'] = df['크기'].map(fruit_size)
```

```python
df.head()
```

```python
# 피벗테이블 (개수와 합계)
pd.pivot_table(df, values = '금액', index =['품목'], columns = ['크기'], aggfunc = ('count', 'sum'))
```

```python
# + 수수료 합계
pd.pivot_table(df, index = ['품목'], columns = ['크기'], aggfunc = {'금액' : ['count', 'sum'], '수수료' : 'sum'})
```
