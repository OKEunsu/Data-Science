## 시간 데잍터 변환

```python
bike_df['datetime'] = pd.to_datetime(bike_df['datetime'])
```

```python
bike_df['year'] = bike_df['datetime'].dt.year
bike_df['month'] = bike_df['datetime'].dt.month
bike_df['hour'] = bike_df['datetime'].dt.hour
```

```python
bike_df['datetime'] = bike_df['datetime'].dt.date
```

```python
plt.figure(figsize=(20,5))
sns.lineplot(x='datetime', y='count', data=bike_df)
plt.show();
```

```python
bike_df[bike_df['year'] == 2019].groupby('month')['count'].mean().to_frame().T
```

```python
bike_df[bike_df['year'] == 2020].groupby('month')['count'].mean().to_frame().T
```

```python
bike_df[bike_df['year'] == 2021].groupby('month')['count'].mean().to_frame().T
```

```python
def covid(date):
  if str(date) < '2020-04-01':
    return 'precovid'
  elif str(date) < '2021-04-01':
    return 'covid'
  else:
    return 'postcovid'
```

```python
bike_df['covid'] = bike_df['datetime'].apply(covid)
```

```python
bike_df.head()
```

```python
bike_df['covid'] = bike_df['datetime'].apply(lambda datetime: 'precovid'
                                             if str(datetime) < '2020-04-01' else 'covid'
                                             if str(datetime) < '2021-04-01' else 'postcovid')
```

```python
bike_df['session'] = bike_df['month'].apply(lambda x:'winter'
                                            if x == 12 else 'fall'
                                            if x >= 9 else 'summer'
                                            if x >= 5 else 'spring'
                                            if x >= 3 else 'winter')
```

```python
bike_df[['session','month']]
```

```python
bike_df['day_night'] = bike_df['hour'].apply(lambda x : 'night'
                      if x >= 21 else 'late evening'
                      if x >= 19 else 'early evening'
                      if x >= 17 else 'late afternoon'
                      if x >= 13 else 'early afternoon'
                      if x >= 9 else 'late morning'
                      if x >= 6 else 'early morning'
                      )
```

```python
bike_df['day_night'].unique()
```

```python
bike_df.head()
```

```python
bike_df.drop(['month', 'datetime'], axis = 1, inplace = True)
```

```python
bike_df.head()
```

## 더미 변수 만들기

```python
bike_df = pd.get_dummies(bike_df, columns = ['weather_main','hour','session', 'covid', 'day_night'])
```

## 모델링

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(bike_df.drop('count', axis = 1), bike_df['count'], test_size = 0.2, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeRegressor

dt = DecisionTreeRegressor(random_state = 42)
dt.fit(X_train, y_train)
```

```python
pred_1 = dt.predict(X_test)
```

```python
sns.scatterplot(x = y_test, y = pred_1)
plt.show();
```

```python
from sklearn.metrics import root_mean_squared_error

root_mean_squared_error(y_test, pred_1)
```

## 선형 회귀와 비교

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train, y_train)
pred_2 = lr.predict(X_test)
```

```python
sns.scatterplot(x = y_test, y = pred_2)
plt.show();
```

```python
root_mean_squared_error(y_test, pred_2)
```

```python
lr.fit(X_train, np.log1p(y_train))
```

```python
pred3 = lr.predict(X_test)
```

```python
pred3 = np.expm1(pred3)
```

```python
root_mean_squared_error(y_test, pred3)
```

## 하이퍼 파라미터 튜닝

```python
dt = DecisionTreeRegressor(random_state = 42, max_depth = 10, min_samples_split = 30)
dt.fit(X_train, y_train)
pred4 =  dt.predict(X_test)
```
