# 결측값 처리

```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'a' : [1, 1, 3, 4, 5], 'b' : [2, 3, np.nan, 3, 4], 'c' : [3, 4, 7, 6, 4]})

# 결측값 확인
df.isnull()

# 결측값 개수 확인
df.isnull().sum()

# 결측값이 포함된 행 지우기
df.dropna()

# 결측값이 포함된 열 지우기
df.dropna(axis=1)

# 결측값을 다른 값으로 대체하기
df = pd.DataFrame({'a' : [1, 1, 3, 4, 5], 'b' : [2, 3, np.nan, 3, 4], 'c' : [3, 4, 7, 6, 4]})

# 특정값으로 대체하기
df.fillna(0, inplace=True)

# 앞이나 뒤의 숫자로 바꾸기
df = pd.DataFrame({'a' : [1, 1, 3, 4, np.nan], 'b' : [2, 3, np.nan, np.nan, 4], 'c' : [np.nan, 4, 1, 1, 4]})

# 1) 뒤의 값으로 채우기
df.fillna(method='bfill')

# 2) 앞의 값으로 채우기
df.fillna(method='ffill')

# limit 설정(위에서 부터 limit로 지정된 갯수만큼만 변경)
df.fillna(method='ffil',limit=1)

# 데이터 프레임에 존재하는 결측값들을 뒤의 값으로 대체한 이후 앞의 값으로 대체하시오.
df.fillna(method='bfill',inplace=True)
df.fillna(method='ffill',inplace=True)

# 평균으로 대체
df = pd.DataFrame({'a' : [1, 1, 3, 4, np.nan], 'b' : [2, 3, np.nan, np.nan, 4], 'c' : [np.nan, 4, 1, 1, 4]})
df.fillna(df.mean()['a']) # a열 평균값으로 대체
df.fillna(df.mean()) # 데이터의 평균값으로 대체

# b,c 의 결측값들을 데이터 프레임의 전체 값의 평균으로 치환하시오.
df.fillna(df.mean()[['b','c']])
```

# apply, map을 활용한 데이터 변환

```python
import pandas as pd

df = pd.DataFrame({'a' : [1, 2, 3, 4, 5]})

# a가 2보다 작으면 2미만, 4보다작으면 4미만, 4보다 크거나 같으면 4이상이라고 추가(apply)
def sort_v(x):
	if x < 2:
		return '2미만'
	elif x <4:
		return '4미만'
	elif x >=4:
		return '4이상'

df['b'] = df['a'].apply(sort_v)
df
```

```python
# a 1 -> one, 2 -> two, 3 -> three, 4 -> four, 5 -> five (map)
a = { 1 : 'one', 2 : 'two', 3 : 'three', 4 : 'four', 5 : 'five'}
df['c'] = df['a'].map(a)
df
```

# 레코드, 칼럼 추가/삭제

```python
import pandas as pd

df = pd.DataFrame({'a' : [1, 1, 3, 4, 5], 'b' : [2, 3, 2, 3, 4], 'c' : [3, 4, 7, 6, 4]})

# 레코드 삭제
df.drop(['d','e','f'], axis =  1 , inplace =  True)

# 레코드 추가
df.append({'a': 6, 'b' : 7, 'c' : 8}, ignore_index =True)

# a에는 7, b에는 8, c에는 9 추가
df.loc[6] = [7,8,9]
df

# 첫번째 레코드  삭제
df.drop(0)

# 첫 번째, 두 번째 레코드를 삭제하시오.
df = df.drop([0,1])

# 첫 번째에서 네 번째 레코드를 삭제하시오.
df.drop(df.index[:4])

# a가 4미만인 레코드들을 삭제하시오.
df[df['a'] < 4].index
df.drop(df[df['a'] < 4].index)

# a가 3미만이고 c가 4인 레코드들을 삭제하시오
df.drop[(df['a'] < 3) & (df['c'] == 4)].index
```
