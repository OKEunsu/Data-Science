# bar plot
- x: x축에 사용될 값
- height: y축에 사용될 값
- width: 각 bar의 width
- bottom: y축의 base 값
- align: 정렬(center or edge)

```python
# bar plot
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6])
plt.show()
```

```python
# width 조정
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], width=0.5)
plt.show()
```

```python
# bottom 조정
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], bottom=2)
plt.show()
```

```python
# align 조정
fig = plt.figure()
ax = fig.add_subplot()
ax.bar([0, 1, 2, 3, 4], [3, 2, 5, 4, 6], align='edge')
plt.show()
```

# histogram plot
- x: 입력 data
- bins: histogram을 나타낼 x축의 구간(정수 입력시 구간의 갯수, list와 같은 sequence 입력시는 sequence 값이 구간으로 설정됨)
- range: x축 범위 (x.min, x.max)로 표현
- density: 확률 분포로 나타낼 것인지 여부

```python
# histogram
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500))
plt.show()
```

```python
# bins
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), bins=50)
plt.show()
```

```python
# bins
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), [-3,-2,-1,0,1,2,3])
# ax.hist(np.random.randn(500), np.linspace(-2, 2, num=40))
plt.show()
```

```python
# range
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), bins=40, range=(-2,2))
plt.show()
```

```python
# density
np.random.seed(1111)

fig = plt.figure()
ax = fig.add_subplot()
ax.hist(np.random.randn(500), bins=40, range=(-2,2), density=True)
plt.show()
```

# Scatter Plot

```python
import pandas as pd
from sklearn.datasets import load_iris

# iris 데이터셋 불러오기
iris = load_iris()
```

```python
iris_df = pd.DataFrame(iris['data'])
iris_df.columns = iris['feature_names']
iris_df['target'] = iris['target']
iris_df.head()
```

```python
fig = plt.figure(figsize=(7,7))

ax = fig.add_subplot()
ax.scatter(x=iris_df['petal length (cm)'],
           y=iris_df['petal width (cm)'])

plt.show()
```

```python
# 3개의 graph 동시에 그리기
fig = plt.figure()

ax = fig.add_subplot()
ax.plot([1, 1.5, 2])
ax.plot([1, 2, 3])
ax.plot([1, 2.5, 4])

plt.show()
```

```python
# line plot과 bar plot 동시에 그리기
fig = plt.figure()

ax = fig.add_subplot()
ax.plot([1, 2, 3], [1, 2, 3])
ax.bar([1, 2, 3], [1, 2, 3])

plt.show()
```

```python
fig = plt.figure(figsize=(7,7))

ax = fig.add_subplot()

for species in iris_df['target'].unique():
  iris_sub = iris_df[iris_df['target']==species]
  ax.scatter(x=iris_sub['petal length (cm)'],
             y=iris_sub['petal width (cm)'])

plt.show()
```

```python
# plt.subplots 활용
fig, axes = plt.subplots(2, 3, figsize=(10,5))
axes[0, 1].plot([1, 2, 3])
axes[1, 2].plot([3, 2, 1])
plt.show()
```

```python
np.random.seed(1111)

fig, axes = plt.subplots(2, 2, figsize=(10,5))
axes[0,0].hist(np.random.randn(200), bins=20)
axes[0,1].bar(np.arange(10), np.random.randn(10)*10+10)
axes[1,0].scatter(np.arange(30), np.arange(30)+3 * np.random.randn(30))
axes[1,1].plot(np.random.randn(50).cumsum()*5+20)

plt.show()
```

```python
# 축 눈금 공유
np.random.seed(1111)

fig, axes = plt.subplots(2, 2, figsize=(10,5), sharey=True)
axes[0,0].hist(np.random.randn(200), bins=20)
axes[0,1].bar(np.arange(10), np.random.randn(10)*10+10)
axes[1,0].scatter(np.arange(30), np.arange(30)+3 * np.random.randn(30))
axes[1,1].plot(np.random.randn(50).cumsum()*5+20)

plt.show()
```

```python
# subplot 간격 조정
np.random.seed(1111)

fig, axes = plt.subplots(2, 2, figsize=(10,5), sharex=True, sharey=True)
for i in range(2):
  for j in range(2):
    axes[i,j].hist(np.random.randn(500), bins=50)

plt.subplots_adjust(wspace=0, hspace=0)

plt.show()
```

```python
np.random.seed(1111)

fig, axes = plt.subplots(2, 2, figsize=(10,5), sharey=True)
axes[0,0].hist(np.random.randn(200), bins=20, color='r')
axes[0,1].bar(np.arange(10), np.random.randn(10)*10+10, color='seagreen')
axes[1,0].scatter(np.arange(30), np.arange(30)+3 * np.random.randn(30), color='#0000ff')
axes[1,1].plot(np.random.randn(50).cumsum()*5+20, color='brown')

plt.show()
```

```python
fig = plt.figure(figsize=(7,7))

ax = fig.add_subplot()

for species in iris_df['target'].unique():
  iris_sub = iris_df[iris_df['target']==species]
  ax.scatter(x=iris_sub['petal length (cm)'],
             y=iris_sub['petal width (cm)'],
             label=species)

plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(7,7))

ax = fig.add_subplot()

for species in iris_df['target'].unique():
  iris_sub = iris_df[iris_df['target']==species]
  ax.scatter(x=iris_sub['petal length (cm)'],
             y=iris_sub['petal width (cm)'],
             label=species)

ax.set_title('Iris')
xticks = np.arange(1, 7.5, 0.5)
ax.set_xticks(xticks)
ax.set_xlabel('PetalLengthCm')
ax.set_ylabel('PetalWidthCm')

plt.legend()
plt.show()
```

```python
import requests

url = 'https://thor.robots.ox.ac.uk/~vgg/data/pets/images.tar.gz'
filename = 'images.tar.gz'

# URL에서 데이터 다운로드
response = requests.get(url)

# 다운로드한 데이터를 파일로 저장
with open(filename, 'wb') as f:
    f.write(response.content)

print("다운로드 완료")
```

```python
import tarfile

filename = 'images.tar.gz'
output_dir = './extracted_images'  # 압축 해제된 파일들이 저장될 디렉토리 경로

# tar.gz 파일 열기
with tarfile.open(filename, 'r:gz') as tar:
    # 압축 해제
    tar.extractall(output_dir)

print("압축 해제 완료")
```
